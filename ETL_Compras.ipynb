Archivo = 'Compras2025.xlsx'

* Tareas:

[ ] Lee los 12 archivos COMPRAS de la carpeta COMPRAS

[ ] Filtra solo filas FC (facturas), excluye PA (pagos)

[ ] Excluye el proveedor EMPRENDIMIENTOS 30002 (ese lo cubre P2/WEG)

[ ] Columnas de salida: fecha, mes, proveedor, concepto, neto

[ ] DataFrame sin duplicados ni filas vacías

[ ] Documentado con docstring y comentarios

Output: df_compras — una fila por factura de compra, 12 meses

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Compras2025.xlsx')

#1. Normalizar nombre de columas

df.columns = (
    df.columns.str.strip().str.lower().str.replace(' ', '_')
)
```

```python
#1. Filtro de columnas mantener FC(facturas) - excluir PA(pagos)
#Ejemplo - excluir 'PA-A-1-1019'

df = df[df['cpte.'].str.startswith('FC')]
df = df[~df['cpte.'].str.startswith('PA')]

df.head(5)
```

```python
#2. Excluir proveedor = EMPRENDIMIENTOS 30002

df = df[df['contacto'] != 'EMPRENDIMIENTOS 30002']

#3. Columnas de salida

df = df[['fecha', 'contacto', 'concepto', 'neto']].copy()

#Ordenar x mes

df['fecha'] = pd.to_datetime(df['fecha'])
df['mes'] = df['fecha'].dt.month

df[['fecha', 'mes']]
```

```python
#4. Eliminar vacíos y duplicados
lista_df = []


df = df.dropna(how = 'all')
lista_df.append(df)

#concatenar

df_compras = pd.concat(
    lista_df,
    ignore_index=True
)

#eliminar duplicados

df_compras = df_compras.drop_duplicates()
df_compras = df_compras.reset_index(drop=True)

df_compras[['fecha', 'mes', 'contacto', 'concepto', 'neto']]   #Output
```

```python
df_compras.to_csv('df_compras.csv', index=False, encoding='utf-8-sig')

# 2. Forzar la descarga automática al escritorio/computadora
from google.colab import files
files.download('df_compras.csv')
```
